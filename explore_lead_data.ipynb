# Lead Scoring Data Exploration & Analysis

This notebook explores your sample lead data to understand structure, patterns, and prepare for LLM model development.

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import os
import glob
from pathlib import Path

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', None)

print("Libraries imported successfully!")
```

## 2. Load Sample Leads Files

```python
data_folder = "sample_data"
data_files = {}

# Find all CSV and Excel files
csv_files = glob.glob(os.path.join(data_folder, "*.csv"))
excel_files = glob.glob(os.path.join(data_folder, "*.xlsx"))

print("=" * 80)
print("📁 DATA FILES FOUND")
print("=" * 80)

# Load CSV files
for csv_file in csv_files:
    file_name = os.path.basename(csv_file)
    print(f"\n📄 Loading: {file_name}")
    try:
        df = pd.read_csv(csv_file)
        data_files[file_name] = df
        print(f"   ✅ Loaded: {len(df)} rows × {len(df.columns)} columns")
    except Exception as e:
        print(f"   ❌ Error: {e}")

# Load Excel files
for excel_file in excel_files:
    file_name = os.path.basename(excel_file)
    print(f"\n📊 Loading: {file_name}")
    try:
        xl_file = pd.ExcelFile(excel_file)
        print(f"   Sheet names: {xl_file.sheet_names}")
        for sheet in xl_file.sheet_names:
            df = pd.read_excel(excel_file, sheet_name=sheet)
            sheet_key = f"{file_name} - {sheet}"
            data_files[sheet_key] = df
            print(f"   ✅ {sheet}: {len(df)} rows × {len(df.columns)} columns")
    except Exception as e:
        print(f"   ❌ Error: {e}")

print(f"\n{'='*80}")
print(f"Total datasets loaded: {len(data_files)}")
print(f"{'='*80}")
```

## 3. Explore Data Structure

```python
for file_name, df in data_files.items():
    print(f"\n{'='*80}")
    print(f"📋 DATASET: {file_name}")
    print(f"{'='*80}")
    print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns\n")
    
    print("Columns & Data Types:")
    print(df.dtypes)
    
    print("\n\nMissing Values:")
    missing = df.isnull().sum()
    if missing.sum() > 0:
        print(missing[missing > 0])
    else:
        print("✅ No missing values")
    
    print("\n\nFirst 3 Rows:")
    print(df.head(3))
```

## 4. Basic Statistics & Data Quality

```python
for file_name, df in data_files.items():
    print(f"\n{'='*80}")
    print(f"📊 STATISTICS: {file_name}")
    print(f"{'='*80}\n")
    
    # Numeric columns
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        print("Numeric Summary:")
        print(df[numeric_cols].describe().round(2))
    
    # Categorical columns
    categorical_cols = df.select_dtypes(include=['object']).columns
    if len(categorical_cols) > 0:
        print("\n\nCategorical Columns (Unique values):")
        for col in categorical_cols:
            print(f"  {col}: {df[col].nunique()} unique values")
            if df[col].nunique() <= 10:
                print(f"    Values: {df[col].unique().tolist()}")
    
    # Data quality
    print(f"\n\nData Quality:")
    print(f"  Total rows: {len(df)}")
    print(f"  Missing values: {df.isnull().sum().sum()}")
    print(f"  Completeness: {(1 - df.isnull().sum().sum() / (len(df) * len(df.columns))) * 100:.1f}%")
```
